# Logistic Regression: Simplified Overview

Logistic regression is a supervised machine learning algorithm used for classification tasks. It predicts the probability of an instance belonging to a specific class (e.g., Class 0 or Class 1). Unlike linear regression, which predicts continuous values, logistic regression is designed for categorical outcomes.

### How It Works
- Logistic regression uses the **sigmoid function** to map input features to a probability value between 0 and 1.
- If the probability exceeds a threshold (commonly 0.5), the instance is classified into one class; otherwise, it is classified into the other.
- It is called "regression" because it builds on the concepts of linear regression but is primarily used for classification problems.

### Example
Imagine predicting whether an email is spam (Class 1) or not (Class 0). Logistic regression analyzes features like the presence of certain keywords, sender information, and email length to calculate the probability of the email being spam.

### Key Features
- **Binary Classification**: Used for tasks with two possible outcomes (e.g., Yes/No, 0/1).
- **Probabilistic Output**: Instead of exact values, it outputs probabilities between 0 and 1.
- **S-Shaped Curve**: Fits a logistic (sigmoid) function instead of a straight line.

---

### Difference Between Linear and Logistic Regression

| **Aspect**              | **Linear Regression**                              | **Logistic Regression**                          |
|--------------------------|----------------------------------------------------|-------------------------------------------------|
| **Purpose**              | Predicts continuous outcomes (e.g., house prices). | Predicts categorical outcomes (e.g., spam or not). |
| **Output**               | Continuous values (e.g., 100, 200).                | Probabilities between 0 and 1.                  |
| **Function Fitted**      | Straight line (best-fit line).                     | S-shaped curve (sigmoid function).              |
| **Use Case**             | Regression tasks (e.g., predicting prices).        | Classification tasks (e.g., binary classification). |

```python
! pip install -q seaborn matplotlib scikit-learn pandas numpy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.datasets import make_classification as mc
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, accuracy_score
```

```python
X, y = mc(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=12)
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)
```

```python
log_reg = LogisticRegression()

log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)
print("Predictions: ", y_pred)
```

```python
# Print the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix: \n", conf_matrix)


# Print the classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report: \n", class_report)

# Print the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

```python
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',cbar=True)
```

# Understanding Confusion Matrix and Classification Metrics

In the context of classification problems, terms like **True Positive (TP)**, **True Negative (TN)**, **False Positive (FP)**, and **False Negative (FN)** are used to evaluate the performance of a model. These terms are derived from the **confusion matrix**, which is a table summarizing the predictions of a classification model.

---

### **Confusion Matrix**
For a binary classification problem, the confusion matrix looks like this:

|                | **Predicted: Positive** | **Predicted: Negative** |
|----------------|--------------------------|--------------------------|
| **Actual: Positive** | True Positive (TP)         | False Negative (FN)        |
| **Actual: Negative** | False Positive (FP)        | True Negative (TN)         |

---

### **Definitions**
1. **True Positive (TP)**:
   - The model correctly predicts the positive class.
   - Example: The model predicts "spam" for an email that is actually spam.

2. **True Negative (TN)**:
   - The model correctly predicts the negative class.
   - Example: The model predicts "not spam" for an email that is actually not spam.

3. **False Positive (FP)**:
   - The model incorrectly predicts the positive class.
   - Example: The model predicts "spam" for an email that is actually not spam.
   - Also known as a **Type I Error**.

4. **False Negative (FN)**:
   - The model incorrectly predicts the negative class.
   - Example: The model predicts "not spam" for an email that is actually spam.
   - Also known as a **Type II Error**.

---

### **Key Metrics Derived from Confusion Matrix**
1. **Accuracy**:
   - Measures the overall correctness of the model.
   - Formula:  
     \[
     \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
     \]

2. **Precision**:
   - Measures how many of the predicted positive cases are actually positive.
   - Formula:  
     \[
     \text{Precision} = \frac{TP}{TP + FP}
     \]

3. **Recall (Sensitivity)**:
   - Measures how many of the actual positive cases are correctly predicted.
   - Formula:  
     \[
     \text{Recall} = \frac{TP}{TP + FN}
     \]

4. **F1-Score**:
   - Harmonic mean of precision and recall, balancing the two.
   - Formula:  
     \[
     \text{F1-Score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
     \]

---

### **Example**
If a spam detection model is evaluated:
- **TP**: Emails correctly classified as spam.
- **TN**: Emails correctly classified as not spam.
- **FP**: Emails incorrectly classified as spam (false alarms).
- **FN**: Emails incorrectly classified as not spam (missed spam).

These metrics help assess the strengths and weaknesses of the model in different scenarios.
