# Logistic Regression using the IRIS Dataset

```python
! pip install -q seaborn openpyxl matplotlib scikit-learn pandas numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, ConfusionMatrixDisplay

```

```python
iris = load_iris()
```

```python
print("IRIS Feature variable : ",iris.feature_names)

print("IRIS Target variables : ",iris.target_names)
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df.iloc[0:1000].to_excel('iris_1.xlsx', index=True)
```

```python
df['Species'] = iris.target
```

```python
df.iloc[0:1000].to_excel('iris_2.xlsx', index=True)
```

```python
df['Species'] = df['Species'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
```

```python
df.iloc[0:1000].to_excel('iris_3.xlsx', index=True)
```

```python
X = df.drop('Species', axis=1)
y = df['Species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python

y_pred = model.predict(X_test)
print(y_pred)
```

```python
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# Using the ConfusionmatrixDisplay Function

# Set the figure size
fig, ax = plt.subplots(figsize=(6,4))
# it will take a default figure size

# Plot the confusion matrix

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap='Blues', ax = ax)
plt.title('Confusion Matrix')
plt.show()
```

```python
# Generate the Classification Report
class_report = classification_report(y_test, y_pred)
print("\nClassification Report",class_report)

Accuracy_score = accuracy_score(y_test, y_pred)
print(f"\nAccuracy Score : {Accuracy_score*100:.2f}%")
```
